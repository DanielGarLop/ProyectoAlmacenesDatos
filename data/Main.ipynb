```python
import pandas as pd

# Cargar datasets (ajusta las rutas según tus archivos)
dataset1 = pd.read_csv("raw/Dataset1.csv")
dataset2 = pd.read_csv("raw/Dataset2.csv")
dataset3 = pd.read_csv("raw/Dataset3.csv")

# Ver estructura básica
print("Dataset1:", dataset1.shape)
print("Dataset2:", dataset2.shape)
print("Dataset3:", dataset3.shape)

# Vista previa
display(dataset1.head())
display(dataset2.head())
display(dataset3.head())
```

```python
def estadisticas_basicas(df, nombre):
    print(f"\n--- Estadísticas de {nombre} ---")
    display(df.describe().T)
    cat_cols = df.select_dtypes(include=['object', 'category']).columns
    for col in cat_cols:
        print(f"\nDistribución de '{col}':")
        print(df[col].value_counts().head(5))

estadisticas_basicas(dataset1, "Dataset1")
estadisticas_basicas(dataset2, "Dataset2")
estadisticas_basicas(dataset3, "Dataset3")
```

```python
import numpy as np

def detectar_outliers(df, nombre):
    print(f"\n--- Outliers en {nombre} ---")
    numeric_cols = df.select_dtypes(include=np.number).columns
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        print(f"Columna {col}: {len(outliers)} posibles outliers")

detectar_outliers(dataset1, "Dataset1")
detectar_outliers(dataset2, "Dataset2")
detectar_outliers(dataset3, "Dataset3")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def graficos_basicos(df, nombre):
    num_cols = df.select_dtypes(include=np.number).columns[:3]
    for col in num_cols:
        plt.figure(figsize=(6,4))
        sns.histplot(df[col], kde=True, bins=20)
        plt.title(f"Distribución de {col} ({nombre})")
        plt.show()

graficos_basicos(dataset1, "Dataset1")
graficos_basicos(dataset2, "Dataset2")
graficos_basicos(dataset3, "Dataset3")
```

```python
import pandas as pd
import numpy as np

# --- FUNCIONES DE LIMPIEZA ---
def limpiar_datos(df):
    print("Iniciando limpieza...")

    # 1️⃣ Eliminar duplicados
    antes = df.shape[0]
    df = df.drop_duplicates().copy()
    print(f"Duplicados eliminados: {antes - df.shape[0]} filas.")

    # 2️⃣ Estandarizar nombres
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

    # 3️⃣ Reemplazar valores vacíos tipo 'NA', etc.
    df = df.replace(['NA', 'N/A', 'null', 'na', 'NaN', 'None'], np.nan)

    # 4️⃣ Manejo de nulos
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna("Desconocido")

    # 5️⃣ Corregir formato de fechas
    for col in df.columns:
        if 'fecha' in col or 'date' in col:
            try:
                df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)
            except Exception:
                pass

    print("✅ Limpieza completada.")
    return df


# --- APLICAR A LOS 3 DATASETS ---
datasets_limpios = {}
for i, df in enumerate([dataset1, dataset2, dataset3], start=1):
    print(f"\n🔹 Limpieza de Dataset{i}")
    datasets_limpios[f"dataset{i}"] = limpiar_datos(df)

print("\n✅ Todos los datasets han sido limpiados.")
```

```python
import os

# Crear la carpeta "data/processed" si no existe
os.makedirs("data/processed", exist_ok=True)

# Guardar cada dataset limpio en formato Excel
for name, df in datasets_limpios.items():
    ruta_salida = f"data/processed/{name}_limpio.xlsx"
    df.to_excel(ruta_salida, index=False)
    print(f"✅ {name} guardado correctamente en {ruta_salida}")
```
